```python
#ceil function to round up a number to its nearest integer
from math import ceil

#function definition
def salestax(itemlist):
    #exempted goods
    exemptedgoods = ['books', 'food', 'medprod']

#initialize    
    oldsum = 0
    newsum = 0
    rnd = 20
    finalOutList = []
    finalprice =[]
    
    for items in itemlist:
        temp_list = items.split()
        
        temp_list = [x.replace(' ','') for x in temp_list]

#presence of exempted goods         
        exempt = False
        
        for item in temp_list:
            if item in exemptedgoods:
                exempt = True
                
        
        if exempt:
            oldsum += float(temp_list[-1])
            priceofone = float(temp_list[-1])/float(temp_list[0])
            
            newprice = (float(temp_list[-1]) * 0.05)+float(temp_list[-1])
            
            newprice = newprice * float(temp_list[0])
            newprice = ceil(round(newprice, 2) * rnd) / rnd
            newsum += newprice
            tempstr = ' '.join(temp_list[:-1])
            print( tempstr+' ' + str(newprice))
            finalprice.append(newprice)

        else:
            if 'imported' in temp_list:
                oldsum += float(temp_list[-1])
                priceofone = float(temp_list[-1])/float(temp_list[0])
                
                newprice = (float(priceofone) * 0.15)+float(priceofone)
                
                newprice = newprice * float(temp_list[0])
                newsum += newprice
                newprice = ceil(round(newprice, 2) * rnd) / rnd
                tempstr = ' '.join(temp_list[:-1])
                print( tempstr+' ' + str(newprice))
                finalprice.append(newprice)
            else:
                oldsum += float(temp_list[-1])
                priceofone = float(temp_list[-1])/float(temp_list[0])
                
                newprice = (float(priceofone) * 0.10)+float(temp_list[-1])
                
                newprice = newprice * float(temp_list[0])
                newsum += newprice
                newprice = ceil(round(newprice, 2) * rnd) / rnd
                tempstr = ' '.join(temp_list[:-1])
                print( tempstr+' ' + str(newprice))
                finalprice.append(newprice)
                
#output 
    print('total:',ceil(round(newsum, 2) * rnd) / rnd)  
    print('salestax:',round((ceil(round(newsum, 2) * rnd) / rnd)-oldsum,2))
           
#input1            
itemlist1 = ['1 book at 12.49', '1 music CD at 14.99', '1 chocolate bar at 0.85']  
salestax(itemlist1)
print('\n')

#input2
itemlist2 = ['1 imported box of food at 10.00','1 imported bottle of perfume at 47.50']    
salestax(itemlist2)  
print('\n')

#input3
itemlist3= [ '1 imported bottle of perfume at 27.99' , '1 bottle of perfume at 18.99','1 packet of headache pills at 9.75', '1 box of imported chocolates at 11.25' ]
salestax(itemlist3)
```
